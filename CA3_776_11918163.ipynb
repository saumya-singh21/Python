# COVID-19 ASSINGNMENT

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:/Users/Lenovo/Desktop/Covid.csv")
data.head()
```

```python
sub_data = data[["Gender","Detected State","Age Bracket"]]
sub_data.head()
```

```python
sub_data.shape
```

```python
sub_data.info()
```

# drop NA

```python
sub_data.dropna(subset = ['Detected State','Gender'],inplace=True)
```

```python
sub_data.head()
```

```python
sub_data["Age Bracket"].fillna('0',inplace=True)
```

```python
def assign_age(x):
    if len(x)>3:
        x1,x2 = x.split('-')
        return str((int(x1)+int(x2))/2).split('.')[0]
    else:
        return x
```

```python
sub_data['Age Bracket'] = sub_data["Age Bracket"].apply(assign_age)
```

```python
sub_data["Age Bracket"] = sub_data['Age Bracket'].astype(int)
```

```python
sub_data.info()
```

```python
age_mean = int(sub_data[sub_data["Age Bracket"]!=0]['Age Bracket'].mean())
```

```python
sub_data["Age Bracket"].replace(0,age_mean,inplace=True)
```

```python
sub_data.head()
```

```python
plt.figure(figsize=(25,10))
plt.xticks(rotation=45)
sns.set(style="white")
sns.barplot(x="Detected State",y='Age Bracket',data=sub_data,hue='Gender',ci=None ,color='b' )
plt.legend(["Female","Male"],prop={'size':15})
plt.xlabel("Affected States",fontdict={"size":15})
plt.ylabel("Average age of Affected Male/Females",fontdict={"size":15})
plt.title("Average Age of affected Males/Females in all states in india",fontdict={"size":25})
```

```python
plt.figure(figsize=(25,10))
sns.set(style="darkgrid")
plt.xticks(rotation=30)
sns.boxplot(x="Detected State",y='Age Bracket',data=sub_data,hue='Gender',color='y',palette="deep")
plt.legend(["Female","Male"],prop={'size':25})
plt.xlabel("Affected States",fontdict={"size":15})
plt.ylabel("Average age of Affected Male/Females",fontdict={"size":15})
plt.title("Average Age of affected Males/Females in all states in india",fontdict={"size":25})
```
